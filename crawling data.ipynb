```python
import nltk
from nltk.tokenize import word_tokenize
import re
import string
```

```python
def hapus_karakter(text):
  text = str(text)
  text = text.replace('\\t', " ").replace('\\n'," ").replace('\\u', " ").replace('\\', "")  #menghapus tab, new line, back slice
  text = text.encode('ascii', 'replace').decode('ascii') #menghapus non ascii
  text = ' '.join(re.sub("([@#_#][A-Za-z0-9]+)|(\w+:\/\/\S+)"," ",text).split())  #menghapus hashtag, mention dan link
  return text.replace("http://", " ").replace("https://", " ")

df['tweet'] = df['tweet'].apply(hapus_karakter)
print(df['tweet'].head())
```

```python
def hapus_punc(text):
    text = str(text)
    return text.translate(str.maketrans('','',string.punctuation))

df['tweet'] = df['tweet'].apply(hapus_punc)
print(df['tweet'].head())
```
